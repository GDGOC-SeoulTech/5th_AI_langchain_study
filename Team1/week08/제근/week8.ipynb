```python
from sklearn.metrics.pairwise import cosine_similarity

# 비교할 문장들
sentence1 = "안녕하세요? 반갑습니다."
sentence2 = "안녕하세요? 반갑습니다!"
sentence3 = "안녕하세요? 만나서 반가워요."
sentence4 = "Hi, nice to meet you."
sentence5 = "I like to eat apples."

sentences = [sentence1, sentence2, sentence3, sentence4, sentence5]

# Gemini 모델을 사용한 임베딩 생성 (유사도 측정을 위해 task_type 지정 가능)
gemini_embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    task_type="semantic_similarity"
)

embedded_sentences = gemini_embeddings.embed_documents(sentences)

def similarity(a, b):
    return cosine_similarity([a], [b])[0][0]

# 결과 출력
print("--- Gemini 임베딩 기반 문장 유사도 ---")
for i in range(len(embedded_sentences)):
    for j in range(i + 1, len(embedded_sentences)):
        score = similarity(embedded_sentences[i], embedded_sentences[j])
        print(f"[유사도 {score:.4f}] {sentences[i]} <=====> {sentences[j]}")
```

```python
from dotenv import load_dotenv
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.storage import LocalFileStore
from langchain.embeddings import CacheBackedEmbeddings

load_dotenv()

# Gemini 임베딩 모델 설정
underlying_embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")

# 로컬 파일 저장소 설정 (임베딩 데이터가 저장될 경로)
store = LocalFileStore("./cache/")

# 캐시 지원 임베딩 생성
cached_embedder = CacheBackedEmbeddings.from_bytes_store(
    underlying_embeddings=underlying_embeddings,
    document_embedding_cache=store,
    namespace=underlying_embeddings.model, # "models/embedding-001" 네임스페이스 사용
)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS

# 1. 문서 로드 및 분할
loader = TextLoader("./data/appendix-keywords.txt")
raw_documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
documents = text_splitter.split_documents(raw_documents)

# 2. 첫 실행: 실제 Gemini API를 호출하여 임베딩하고 캐시에 저장
print("--- 첫 번째 실행 (API 호출 포함) ---")
%time db = FAISS.from_documents(documents, cached_embedder)

# 3. 두 번째 실행: API 호출 없이 로컬 캐시에서 즉시 로드
print("\n--- 두 번째 실행 (로컬 캐시 사용) ---")
%time db2 = FAISS.from_documents(documents, cached_embedder)
```

```python
from langchain_ollama import OllamaEmbeddings
import numpy as np

# 1. 로컬 Ollama 임베딩 모델 초기화
# 별도의 API 키가 필요 없습니다.
embeddings = OllamaEmbeddings(
    model="nomic-embed-text", # 혹은 "bge-m3" 등 사용 가능
)

texts = [
    "안녕, 만나서 반가워.",
    "LangChain simplifies the process of building applications with large language models",
    "랭체인 한국어 튜토리얼은 LangChain의 실용 예제를 바탕으로 합니다.",
    "LangChain은 초거대 언어모델로 애플리케이션을 구축하는 과정을 단순화합니다.",
    "Retrieval-Augmented Generation (RAG) is an effective technique."
]

# 2. 문장 임베딩 생성 (로컬 CPU/GPU 자원 사용)
embedded_documents = embeddings.embed_documents(texts)
embedded_query = embeddings.embed_query("LangChain에 대해 알려줘.")

# 3. 유사도 계산 (코사인 유사도 활용 권장)
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# 결과 계산 및 정렬
scores = [cosine_similarity(embedded_query, doc) for doc in embedded_documents]
sorted_results = sorted(zip(scores, texts), key=lambda x: x[0], reverse=True)

# 4. 결과 출력
print(f"[Query] LangChain에 대해 알려줘. (Local Ollama 기반)")
print("=" * 50)
for score, text in sorted_results:
    print(f"[{score:.4f}] {text}")
```

```python
import os

# 1. data 폴더가 없으면 생성
os.makedirs("data", exist_ok=True)

# 2. 테스트용 파일 생성 (파일이 없을 경우를 대비)
with open("data/nlp-keywords.txt", "w", encoding="utf-8") as f:
    f.write("자연어 처리(NLP)는 인간의 언어를 이해하는 기술입니다.")

with open("data/finance-keywords.txt", "w", encoding="utf-8") as f:
    f.write("금융 데이터 분석은 주가 예측과 리스크 관리에 중요합니다.")

print("폴더와 파일이 준비되었습니다.")
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter # 경로 변경됨
from langchain_chroma import Chroma

# 1. 문서 로더 설정 (여러 파일을 리스트로 관리하면 확장성이 좋습니다)
file_paths = ["data/nlp-keywords.txt", "data/finance-keywords.txt"]
docs = []

for path in file_paths:
    loader = TextLoader(path, encoding='utf-8') # 인코딩 명시 권장
    docs.extend(loader.load())

# 2. 텍스트 분할기 설정
# langchain.text_splitter 대신 langchain_text_splitters를 사용하는 것이 최신 표준입니다.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=0,
    add_start_index=True # 각 청크의 시작 위치를 메타데이터에 포함 (디버깅 용이)
)

# 3. 문서 분할 수행
# load_and_split() 보다는 load() 후 split_documents()를 사용하는 것이 로직 분리에 유리합니다.
split_docs = text_splitter.split_documents(docs)

# 4. 결과 확인
print(f"전체 분할된 문서 개수: {len(split_docs)}")

# (참고) 이후 벡터 스토어 생성 시 예시
# embeddings = OpenAIEmbeddings(model="text-embedding-3-small") # 최신 모델 사용
# vectorstore = Chroma.from_documents(documents=split_docs, embedding=embeddings)
```

```python
# DB 생성
db = Chroma.from_documents(
    documents=split_docs, embedding=OpenAIEmbeddings(), collection_name="my_db"
)
```

```python
# 1. 패키지 경로 변경 (v0.3 표준)
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os

# 2. 텍스트 분할기 설정 (분리된 패키지 사용)
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=0,
    add_start_index=True  # 각 청크의 시작 위치를 메타데이터에 기록 (권장)
)

# 3. 여러 파일을 효율적으로 로드하고 분할하는 현대적인 방식
file_paths = ["data/nlp-keywords.txt", "data/finance-keywords.txt"]
all_split_docs = []

for path in file_paths:
    if os.path.exists(path):
        loader = TextLoader(path, encoding="utf-8")
        # load()와 split_documents()를 명시적으로 분리하는 것이 디버깅과 유지보수에 유리합니다.
        docs = loader.load()
        split_docs = text_splitter.split_documents(docs)
        all_split_docs.append(split_docs)
        print(f"{path}: {len(split_docs)} chunks 생성됨")
    else:
        print(f"파일 없음: {path}")

# 개별 확인 (기존 요청하신 형태)
# split_doc1, split_doc2 = all_split_docs[0], all_split_docs[1]
```
